```python
#esikäsittely
import pandas as pd

url = 'https://student.labranet.jamk.fi/~huoptu/datananal_s20/kerta5/harjoitukset/autot.csv'
df1 = pd.read_csv(url, sep='\t', engine='python')

#suodatetaan mukaan vain yli 50 auton valmistajat
mh_counter = df1['Mh'].value_counts()
df2 = df1[df1['Mh'].isin(mh_counter.index[mh_counter > 50])].copy()

#uudelleennimetään polttoaineet
d = {'DIESEL': 'Diesel', 'PETROL': 'Petrol', 'NG': 'Natural gas'}
df2.replace({'Ft': d}, inplace=True)

#Tässä tehtävässä ei tehdäkään crosstabia, koska countplot osaa laskea suoraan!
```

```python
#hieno horisontaalinen pylväsdiagrammi
import matplotlib.pyplot as plt
import seaborn as sns

ax = sns.countplot(y=df2['Mh'], hue=df2['Ft'], data=df2, edgecolor='black')
ax.set_title(label='Number of cars by fuel type')
ax.set_ylabel('Manufacturer name' )
ax.set_xlabel('Number of cars' )
ax.legend(title='Fuel type', loc='lower right')
ax.grid(color='black', linestyle='--', axis='both')
```
