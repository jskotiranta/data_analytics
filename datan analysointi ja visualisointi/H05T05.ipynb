```python
#esikäsittely
import pandas as pd

url = 'https://student.labranet.jamk.fi/~huoptu/datananal_s20/kerta5/harjoitukset/autot.csv'
df1 = pd.read_csv(url, sep='\t', engine='python')

#suodatetaan mukaan vain yli 50 auton valmistajat ja autot
#joiden päästöt ovat suuremmat kuin nolla
mh_counter = df1['Mh'].value_counts()
df2 = df1[df1['Mh'].isin(mh_counter.index[mh_counter > 50])].copy()
df = df2[(df2['Enedc (g/km)'] > 0)].copy()
```

```python
#kertakaikkisen huikea visualisointi
import matplotlib.pyplot as plt
import seaborn as sns

ax1 = sns.histplot(data=df['Enedc (g/km)'], bins=100, stat='density')
ax1.set(yscale='log')
ax1.set_ylabel('Probability density' )
ax1.set_xlabel('Specific $CO_2$ Emissions, (g/km)' )

ax2 = sns.ecdfplot(data=df['Enedc (g/km)'], color='red')
ax2.legend(['Cum. specific $CO_2$ Emissions', 'Specific $CO_2$ Emissions'], loc='lower right')
plt.show() #tulipa selkeä ja professionaalin näköinen :D
```
